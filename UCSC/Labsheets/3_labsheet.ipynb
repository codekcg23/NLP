## More Specialized Frequency Distributions(What is a word?)

```python
# Removing special characters using regular expression
import re
pattern = re.compile('.*[^a-z].*')
nonAlphaMatch = pattern.match('-')
if nonAlphaMatch: 
    print("matched non alphabetical")
```

```python
stopwords=['to','be','of','the','in','it','was','i','am','she','had','been','is','have','could','not','her','he','do','and','would','such','a','his','must']
word ='the'
if word in stopwords:
    print("Stop")
```

```python
def alphaStopFreqDist(words,stoplist):
    #make a new frequency distribution called asdist 
    asdist=FreqDist()
    #define the regular expression pattern to match
    #non-alphabeticaltokens
    pattern=re.compile('.*[^a-z].*')
    #for every token,if it doesn't match the non-alphabetical #pattern and if it is not on the stop word list add it to 
    #the frequency distribution 
    for word in words:
        if not pattern.match(word):
            if not word in stoplist:
                asdist[word] +=1              #returnthefrequencydistributionastheresult
    return asdist
```

```python
asdist = alphaStopFreqDist(shortwords,stopwords)
asdist.keys()
```

```python
for key in asdist.keys():
    print(key,asdist[key])
```

```python
bigasdist = alphaStopFreqDist(emmawords,stopwords)
bigasdist.keys()
```

```python
for key in bigasdist.keys():
    print(key,bigasdist[key])
```
